```python
import pandas as pd
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import graphviz
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
df = pd.read_csv('diabetes.csv') 
```

```python
task_data = df.head(670)
```

**Первый ответ**

```python
len(task_data[task_data['Outcome'] == 0])
```

```python
train = task_data.head(int(len(task_data)*0.8))
test = task_data.tail(int(len(task_data)*0.2))
```

```python
features = list(train.columns[:8])
x = train[features]
y = train['Outcome']
```

```python
tree = DecisionTreeClassifier(criterion='entropy', #критерий разделения
                              min_samples_leaf=10, #минимальное число объектов в листе
                              max_leaf_nodes=10, #максимальное число листьев
                              random_state=2020)
clf=tree.fit(x, y)
```

**Второй ответ**

```python
clf.tree_.max_depth
```

```python
columns = list(x.columns)
export_graphviz(clf, out_file='tree.dot', 
                feature_names=columns,
                class_names=['0', '1'],
                rounded = True, proportion = False, 
                precision = 2, filled = True, label='all')

with open('tree.dot') as f:
    dot_graph = f.read()

graphviz.Source(dot_graph)
```

_Доля правильных ответов_

```python
features = list(test.columns[:8])
x = test[features]
y_true = test['Outcome']
y_pred = clf.predict(x)
accuracy_score(y_true, y_pred)
```

**Шестой ответ**

```python
f1_score(y_true, y_pred, average='macro')
```

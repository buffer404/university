## Упражнение 2.1

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn import datasets
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

**Значение 1** это первый ответ  
**Значение 2** это второй ответ

```python
X = pd.read_csv("14_25.csv", header = None)
pca = PCA(n_components=2, svd_solver='full')
X_transformed = pca.fit(X).transform(X)
X_transformed[0]
```

Пример:  
    _Введите долю объясненной дисперсии при использовании первых двух главных компонент._
    Мой ответ 0.781
    
Если у вас 3 элемента - первое значение
2 элемента - второе значение

```python
explained_variance = np.round(np.cumsum(pca.explained_variance_ratio_),3)
explained_variance
```

Пример:  
_Какое минимальное количество главных компонент необходимо использовать, чтобы доля объясненной дисперсии превышала 0.85_ Мой ответ 3

Нумерация идет с единицы

```python
pca = PCA(n_components=10, svd_solver='auto')
X_full = pca.fit(X).transform(X)
explained_variance = np.round(np.cumsum(pca.explained_variance_ratio_),3)
explained_variance
```
